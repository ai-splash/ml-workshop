```python
#euclidean distance
def euclidean_distance(point1,point2):
    dist = np.linalg.norm(point1-point2)
    return dist
#euclidean_distance(np.array([1,2]),np.array([2,1]))
```

#  Euclidean Distance

```python
def euclidean_distance_for_single_point(X_train,y_train, point):
#     single thanda getheyyyyyy !!!!!
#     singlism
#     vaalga singles
#     vinod arjun,praveen benedict is pacha single daa 
#     thilak god is universal playboy
    total_distance=[]

    for i,x in enumerate(X_train):
        distance=np.linalg.norm(x-point)
        #print(i,x)
    
        total_distance.append((distance,y_train[i]))
    return total_distance
    #return [np.linalg.norm(X - point) for X in X_train]
```

# Memorize the data

```python
def memorize(X_train,y_train,X_test):
    memorized_results=[]
    for x in X_test:
        #single tha getheyyyy!!
        single_point_distance=euclidean_distance_for_single_point(X_train,y_train,x)
#         print(len(single_point_distance))
        memorized_results.append(single_point_distance)
    return memorized_results
        
```

# get the k neighbours

```python
from collections import Counter
def sort_and_slice(distance_list,k):
    vote=0
    count=0  
    sorted_distance_list=[]
    for element in distance_list:
        sorted_element=sorted(element,key=lambda x:x[0])
        elements=sorted_element[0:k]
        sorted_distance_list.append(elements)
    return sorted_distance_list
#dankmemes generation mc gaali
#gan automation avr gaali
#web automation yoga gaali
```

# Voting 

```python
def find_mode(k_list):
    result=[]
    array=[]
    for element in k_list:
        classlist=[]
        for tuple in element:
        #print(tuple[1])
            classlist.append(tuple[1])
        array.append(classlist)
    for i in array:
    #print(i)
        data=Counter(i)
        a=data.most_common(1)
        #print(a[0][0])
        result.append(a[0][0])
    return result
            
```

# Predict

```python
def predict(X_train,y_train,X_test,k):
  
    distance_list = memorize(X_train,y_train,X_test)
    #print(distance_list)
    k_list=sort_and_slice(distance_list,k)
    #find mode of class
    #print(k_list)

    final_results=find_mode(k_list)
    #print(len(distance_list))
    #print(k_list)
    return np.array(final_results)
    
```

# Optical Character Recognition

```python
from sklearn import datasets
#from skimage import exposure
from sklearn.cross_validation import train_test_split
```

# Loading the mnists

```python
mnist = datasets.load_digits()
```

```python
data=mnist["images"]
labels=mnist["target"]
```

# data visualization

```python
i=432
plt.imshow(data[i],cmap="binary")
print("this image is handwriiten digit of "+str(labels[i]))
```

# Train test datasplit

```python
(X_train, X_test,y_train , y_test) = train_test_split(data,labels, test_size=0.25, random_state=42)
```

```python
print("training data points: {}".format(len(y_train)))
print("testing data points: {}".format(len(y_test)))
```

# Flattening the image as single vector

```python
no_of_training_examples=X_train.shape[0]
height=X_train.shape[1]
width=X_train.shape[2]
X_train_flatten=X_train.reshape(no_of_training_examples,height*width)
```

```python
X_test_flatten=X_test.reshape(X_test.shape[0],height*width)
X_test_flatten.shape
```

# Data transformation / normalization


```python
X_train_flatten/=255
X_test_flatten/=255
```

# Working with the test data

```python
plt.imshow(X_test[13].reshape(8,8),cmap="binary")
```

```python
y_pred=predict(X_train_flatten,y_train,X_test_flatten,k=3)
print("the predicted class of given handwriiten pic by kNN is\t===>\t{}".format(y_pred[13]))
```

# Accuracy (Model evaluation)

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test,y_pred )
score*100
```
